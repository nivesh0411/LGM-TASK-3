```python
# Importing The Libraries
import cv2
from google.colab.patches import cv2_imshow

```

```python
# Loading & Displaying The RGB Image
image=cv2.imread("/content/Demonslayer.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
# Converting The RGB Image To Grayscale Image
img_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
cv2.waitKey(0)
```

```python
# Inverting The Grayscale Image
img_invert=255-img_gray
cv2_imshow(img_invert)
cv2.waitKey(0)
```

```python
# Blurring The Inverted Image Using Gaussian Function
img_blur=cv2.GaussianBlur(img_invert,(21,21),0)
```

```python
# Inverting The Blurred Image
blur_invert=255-img_blur
sketch_pencil=cv2.divide(img_gray,blur_invert,scale=256.0)
cv2_imshow(sketch_pencil)
cv2.waitKey(0)
```

```python
# Comparing the Original RGB Image & Pencil-Sketched Image
print("Original RGB Image:\n")
cv2_imshow(image)
print("\nPencil-Sketched Image Usimg OpenCV:\n")
cv2_imshow(sketch_pencil)
cv2.waitKey(0)
```
